# 2)
y1=√x \
y2=x3 \
y3=2x \
y4=1/(x+1)

```python
x = list(range(-150,150))
y1=[]
y2=[]
y3=[]
y4=[]

for _ in x:
	y1.append((_**(1/2)).real)
	y2.append(((_*0.5)**3).real)
	y3.append((2**_).real)
	if _==-10:
		y4.append(None)
		continue
	else:
		y4.append((1/((_*0.1)+1)).real)

# y=[(2**_) for i in x]	

plt.plot(x, y1, c='red', alpha=0.7)
plt.xlim(-5, 150)
plt.ylim(-5, 20)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=sqrt(x)')
plt.show()

plt.plot(x,y2, c='red', alpha=0.7)
plt.xlim(-10, 10)
plt.ylim(-20, 20)
plt.text(0, -2, '(0, 0)', fontsize=10, color='blue')
plt.title('y=x**3')
plt.show()

plt.plot(x,y3, c='red', alpha=0.7)
plt.xlim(-75, 50)
plt.ylim(-5, 100)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=2**x')
plt.show()

plt.plot(x,y4, c='red', alpha=0.7)
# plt.xlim(-2, 0)
# plt.ylim(-20, 20)
plt.text(-15, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=1/(x+1)')
plt.show()

```

# 3
2 from above, but sublimed

```python

x = list(range(-150,150))
y1=[]
y2=[]
y3=[]
y4=[]

for _ in x:
	y1.append((_**(1/2)).real)
	y2.append((_**3).real)
	y3.append((2**_).real)
	if _==-1:
		y4.append(None)
		continue
	else:
		y4.append((1/(_+1)).real)

def four_subplots(x,y1,y2,y3,y4):
	plt.figure(figsize=(16, 9))
	n_rows = 2
	n_cols = 2

	plt.subplot(n_rows, n_cols, 1)
	plt.plot(x, y1)
	plt.title('x ~ y1')

	# the second subplot
	plt.subplot(n_rows, n_cols, 2)
	plt.plot(x, y2)
	plt.title('x ~ y2')

	plt.subplot(n_rows, n_cols, 3)
	plt.plot(x, y3)
	plt.title('x ~ y3')

	# the second subplot
	plt.subplot(n_rows, n_cols, 4)
	plt.plot(x, y4)
	plt.title('x ~ y4')

	return plt.show()

print(four_subplots(x,y1,y2,y3,y4))
```
