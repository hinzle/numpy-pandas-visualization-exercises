# 2)
y1=âˆšx \
y2=x3 \
y3=2x \
y4=1/(x+1)

```python
x = list(range(-150,150))
y1=[]
y2=[]
y3=[]
y4=[]

for _ in x:
	y1.append((_**(1/2)).real)
	y2.append((_**3).real)
	y3.append((2**_).real)
	if _==-1:
		y4.append(None)
		continue
	else:
		y4.append((1/(_+1)).real)

plt.plot(y1, x, c='red', alpha=0.7)
plt.xlim(-5, 25)
plt.ylim(-5, 125)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=sqrt(x)')
plt.show()

plt.plot(y2, x, c='red', alpha=0.7)
plt.xlim(-175, 175)
plt.ylim(-10, 10)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=x**3')
plt.show()

plt.plot(y3, x, c='red', alpha=0.7)
plt.xlim(-75, 175)
plt.ylim(-5, 10)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=2**x')
plt.show()

plt.plot(y4, x, c='red', alpha=0.7)
plt.xlim(-10, 10)
plt.ylim(-25, 25)
plt.text(5, 0, '(0, 0)', fontsize=10, color='blue')
plt.title('y=1/(x+1)')
plt.show()

```
